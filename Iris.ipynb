```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='whitegrid')
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('Iris.csv')
data.head()
```

```python
sns.pairplot(data, hue = 'Species')
```

```python
# kde plot of SepalLengthCm versus SepalWidthCm for Iris-setosa species of flower
setosa = data[data['Species']=='Iris-setosa']
sns.kdeplot( setosa['SepalWidthCm'], setosa['SepalLengthCm'],
                 cmap="viridis", shade=True, shade_lowest=False)
```

```python
# Importing train test split
from sklearn.model_selection import train_test_split
```

```python
X  = data.drop(['Id', 'Species'], axis = 1)
y = data['Species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Importing Support Vector Machine model for Classification
from sklearn.svm import SVC
```

```python
svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
# Importing metrics from sklearn for confusion matrix and classificaton report
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, y_pred))
print('\n')
print(classification_report(y_test, y_pred))
```
